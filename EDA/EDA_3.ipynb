```python
months = ['january', 'february', 'march', 'april', 'may', 'june', 'july',
          'august', 'september', 'october', 'november', 'december']
buildings_woTemp= buildings.copy()
for key in buildings.keys():
    if (key.split('_')[0]) in months:
        # print(key)
        buildings_woTemp= buildings_woTemp.drop(key,axis=1,inplace=False)
display(buildings_woTemp)
```

# Correlation

```python
corrm = buildings.corr()
corrm['site_eui']
pos_cor = corrm['site_eui'] >0.1
corm_pos= corrm['site_eui'][pos_cor].sort_values(ascending = False).to_frame()
display(corm_pos)
neg_cor = corrm['site_eui'] <-0.1
corm_neg=corrm['site_eui'][neg_cor].sort_values(ascending = True).to_frame()
display(corm_neg)
```
